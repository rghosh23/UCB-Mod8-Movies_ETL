```python
# first load wiki_movies into a dataframe
wiki_movies_df = pd.DataFrame(wiki_movies_raw)
wiki_movies_df.head()
```

```python
# list all column heads
wiki_movies_df.columns.tolist()
```

```python
# use list comprehension to filter wiki JSON file to only include entries with a director and an IMDB link (using for)
wiki_movies = [movie for movie in wiki_movies_raw
               if ('Director' in movie or 'Directed by' in movie)
                   and 'imdb_link' in movie]
len(wiki_movies)
```

```python
# add another filter to remove TV shows since there arent too many of those
wiki_movies = [movie for movie in wiki_movies_raw
               if ('Director' in movie or 'Directed by' in movie)
                   and 'imdb_link' in movie
                   and 'No. of episodes' not in movie]
len(wiki_movies)
```

```python
# use function clean_movie to clean movie data
```

```python
# start function with creating a destructive copy
#def clean_movie(movie):
#    movie_copy = dict(movie)
```

```python
#Step 1
def clean_movie(movie):
    movie = dict(movie) #create a non-destructive copy
    # Step 2: clean language columns
    # 2a- dictionary to hold alternate titles
    alt_titles = {}
    # 2b- loop through all the list of alt titles
    for key in ['Also known as','Arabic','Cantonese','Chinese','French',
                'Hangul','Hebrew','Hepburn','Japanese','Literally',
                'Mandarin','McCune–Reischauer','Original title','Polish',
                'Revised Romanization','Romanized','Russian',
                'Simplified','Traditional','Yiddish']:
        #2c- check if current key is in movie
        if key in movie:
            alt_titles[key] = movie[key]
            movie.pop(key)
    #2d- add alt titles to movie
    if len(alt_titles) > 0:
        movie['alt_titles'] = alt_titles 
    
    return movie
```

```python
# after step 2, make a list of clean movies with list comprehension
clean_movies = [clean_movie(movie) for movie in wiki_movies]
```

```python
# create df from clean movies
wiki_movies_df = pd.DataFrame(clean_movies)
sorted(wiki_movies_df.columns.tolist())
```
